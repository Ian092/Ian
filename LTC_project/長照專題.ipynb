```python
import pandas as pd
file_path = '/content/drive/MyDrive/python/趙雲瀚_AI應用設計/長照專題/修改後的統計表.csv'
data = pd.read_csv(file_path)
data.head()
```

##分割數據集

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 特徵
features = data[['地區人口數', '老年人口數', '長照需求人數', '老化指數', '扶老比', '就業人口', '薪資中位數', '醫療院所數', '照服人力']]
target = data['長照機構數']

# 分割數據集
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 訓練決策樹回歸模型
model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)

# 測試數據進行預測
y_pred = model.predict(X_test)

# 評估模型性能
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, r2
```

```python
# 使用訓練好的模型對當前數據進行預測
predicted_values = model.predict(features)

# 將實際值和預測值存入數據框架中進行比較
results = data[['地區', '長照機構數']].copy()
results['預測長照機構數'] = predicted_values

# 顯示數據框架的前幾行
print(results.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree


file_path = '/content/drive/MyDrive/python/趙雲瀚_AI應用設計/長照專題/修改後的統計表.csv'
data = pd.read_csv(file_path)

# 特徵
features = data[['地區人口數', '老年人口數', '長照需求人數', '老化指數', '扶老比', '就業人口', '薪資中位數', '醫療院所數', '照服人力']]
target = data['長照機構數']

# 分割數據集
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 訓練決策樹回歸模型
model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)

plt.figure(figsize=(20, 10))

# 決策樹
plot_tree(model, feature_names=features.columns, filled=True, rounded=True, fontsize=10)

plt.title("長照機構數的決策樹")
plt.show()
```
